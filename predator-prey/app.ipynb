```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Parameters
##### Simulation with 500 hares and 10 foxes

```python
hares = 500                             #amount of hares
foxes = 10                              #amount of foxes
delta_t = 1                             #number of weeks per cycle        
increase_foxes = 0.0004                 #fox growth rate per encounter
hares_increase_rate_per_week = 0.08     #rate of increase of hares per week
death_rate_foxes = 0.2                  #death rate of foxes without food
hares_loss_rate= 0.002                  #hares loss rate per encounter
initial_time = 0                        #initial time
final_time = 500                        #final time (number of cycles)
ground_capacity = 1400                  #total land capacity
ground_capacity_to_feed = 0             #land capacity to feed hares
increase_hares = 0                      #increase of hares
```

```python
def scenario_population(hares_list, foxes_list):
    df = pd.DataFrame({'hares': hares_list,
                   'foxes': foxes_list})
    ax=df.plot(kind='line', y='hares', color='DarkBlue')
    ax2=df.plot(kind='line', y='foxes', secondary_y=True, color='Red', ax=ax)
    ax.set(xlabel='Time (weeks)', ylabel='Hares Population', title='Scenario Population')
    ax2.set(xlabel='Tieme (weeks)', ylabel='Foxes Population')
    plt.show()

def diagram_phase(hares_list, foxes_list):
    df = pd.DataFrame({'hares': hares_list,
                   'foxes': foxes_list})
    ax=df.plot(kind='line', x='hares',y='foxes', color='DarkBlue')
    ax.set(ylabel='Foxes Population', xlabel='Hares Population', title='Phase Diagram')
    
    leg = plt.legend()
    ax.get_legend().remove()
    plt.show()

foxes_list = []
hares_list = []
    
def main(hares, foxes, delta_t, increase_foxes, hares_increase_rate_per_week, death_rate_foxes, hares_loss_rate, final_time, ground_capacity):
    initial_hares = hares
    initial_foxes = foxes
    for i in range(final_time):
        ground_capacity_to_feed = ground_capacity - hares
        increase_hares = (1/ground_capacity) * ground_capacity_to_feed * hares_increase_rate_per_week * hares
        foxes_death = death_rate_foxes * foxes
        encounter = hares * foxes
        hares = hares + (delta_t * (increase_hares - hares_loss_rate * encounter))
        foxes = foxes + (delta_t * (increase_foxes * encounter - foxes_death))
        hares_list.append(hares)
        foxes_list.append(foxes)
    scenario_population(hares_list, foxes_list)
    diagram_phase(hares_list, foxes_list)   
    
    print(f"-Initial fox population: {(initial_foxes)}\n-Initial hare population: {int(initial_hares)}\n**************************")
    print(f"-Fox population at mid-cycle: {(foxes_list[int(final_time/2)])}\n-Hare population at mid-cycle: {int(hares_list[int(final_time/2)])}\n**************************")
    print(f"-Final fox population: {(foxes)}\n-Final hare population: {int(hares)}\n**************************")
    print(f"-Maximum fox population: {(max(foxes_list))}\n-Minimum fox population: {int(min(foxes_list))}\n**************************")
    print(f"-Maximum hare population: {(max(hares_list))}\n-Minimum hare population: {int(min(hares_list))}\n**************************")
    
if __name__ == "__main__":
    main(hares, foxes, delta_t, increase_foxes, hares_increase_rate_per_week, death_rate_foxes, hares_loss_rate,final_time, ground_capacity)
```

## ANALYSIS
This simulation starts with 500 hares and 10 foxes, cycling every week, for 500 weeks, with a fox growth rate per encounter of 0.0004, with a fox death rate without food of 0.2, and a rate of 0.08 increase of hares per week and a loss rate of hares per encounter of 0.002 and a field capacity of 1400.

It is observed that the fox population had a large increase reaching a maximum of 43 foxes, leaving a population of 380 hares, from there they oscillated until reaching a point of equilibrium, with a population of 25 foxes and 500 hares after 500 weeks. Comparing from the initial amount to the final amount, we can say that the fox population increased and the number of hares returned approximately to the initial value.

## ANALISIS
Esta simulación comienza con 500 liebres y 10 zorros, ciclando cada una semana, durante 500 semanas, con una tasa de crecimiento de zorros por encuentro del 0,0004, con una tasa de mortalidad de zorros sin alimentos del 0,2 y una tasa del 0,08 de incremento de liebres por semana y una tasa de perdida de liebres por encuentro del 0,002 y una capacidad del terreno de 1400.
